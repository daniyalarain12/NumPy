# NumPy ---> STANDS FOR NUMERICAL PYTHON
- NumPy is a Python library used for numerical computing.
- It provides a high-performance multidimensional array (ndarray) and a wide range of functions to perform fast      mathematical operations on large datasets efficiently.

## DIFFERENCE BETWEEN Python Lists AND NumPy Array

| Feature           | Python List         | NumPy Array        |
| ----------------- | ------------------- | ------------------ |
| Memory Storage    | Non-contiguous      | **Contiguous**     |
| Data Type         | Mixed types allowed | Same type required |
| Speed             | Slower              | Faster             |
| Memory Efficiency | Less efficient      | More efficient     |

# PYTHON LISTS VS NUMPY ARRAYS

```python
import numpy as np
import time
import sys
```

## EXECUTION TIME COMPARISON

```python
size = 100_00_00

# PYTHON LISTS
python_list = list(range(size))
start = time.time()
square_list = [i*i for i in python_list]
end = time.time()
print(f"EXECUTION TIME OF PYTHON LISTS : {end-start} SECONDS")

# NUMPY ARRAYS
arr = np.array(python_list)
start = time.time()
square_array = arr ** 2
end = time.time()
print(f"EXECUTION TIME OF NUMPY ARRAYS : {end-start} SECONDS")
```

## MEMORY USAGE COMPARISON

```python
# PYTHON LISTS
print(f"SIZE OF PYTHON LISTS : {sys.getsizeof(python_list) * len(python_list)} BYTES")

# NUMPY ARRAYS
print(f"SIZE OF NUMPY ARRAYS : {arr.nbytes} BYTES")
```
