## Exploring Attributes of a One-Dimensional Array

- Create a one-dimensional NumPy array.
- Examine its dimension, shape, data type, and number of elements.
- Calculate the total memory occupied by the array using both manual calculation and the nbytes attribute.

```python
import numpy as np
```

```python
a = np.array([1,2,3])

print("a =",a)
print("DIMENSION OF a IS:",a.ndim)
print("SHAPE OF a IS:",a.shape)
print("DATA TYPE OF a:",a.dtype)
print("SIZE OF a:",a.itemsize,"BYTES")
print("TOTAL NO. OF ELEMENTS IN a:",a.size)
print("TOTAL SIZE OF a:",a.size * a.itemsize,"BYTES")
print("TOTAL SIZE OF a:",a.nbytes,"BYTES")
```

## Exploring Attributes of a Two-Dimensional Array

- Create a two-dimensional NumPy array with a custom data type.
- Examine its dimension, shape, data type, and number of elements.
- Calculate the total memory occupied by the array using both manual calculation and the nbytes attribute.

```python
b = np.array([[1,2,3],[4,5,6]], dtype='int16')

print("b =",b)
print("DIMENSION OF b IS:",b.ndim)
print("SHAPE OF b IS:",b.shape)
print("DATA TYPE OF b:",b.dtype)
print("SIZE OF b:",b.itemsize,"BYTES")
print("TOTAL NO. OF ELEMENTS IN b:",b.size)
print("TOTAL SIZE OF b:",b.size * b.itemsize,"BYTES")
print("TOTAL SIZE OF b:",b.nbytes,"BYTES")
```

## Changing the Data Type of a NumPy Array

- Use astype() to convert values from one data type to another.
- Convert a floating-point array into an integer array.

```python
c = np.array([23.5,45.6,67.5])

print("c =",c)
print("DATA TYPE OF c:",c.dtype)

int_arr = c.astype(int)

print("c =",int_arr)
print("DATA TYPE OF c:",int_arr.dtype)
```
