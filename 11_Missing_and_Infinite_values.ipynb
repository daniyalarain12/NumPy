# NumPy Handling Missing and Infinite Values

- Learn how to detect missing values in NumPy arrays.
- Understand how to replace missing values with custom values.
- Learn how to identify positive and negative infinite values.
- Replace infinite values to prepare data for analysis and machine learning tasks.

## Detecting Missing Values with isnan()

- Use isnan() to identify missing values in an array.
- The function returns True where a missing value (NaN) is present.
- NaN stands for Not a Number.

```python
import numpy as np
```

```python
arr = np.array([1,2,np.nan,3,4,np.nan,5])

print(np.isnan(arr))
```

## Replacing Missing Values with nan_to_num()

- Use nan_to_num() to replace missing values in an array.
- By default, NaN values are replaced with 0.
- A custom replacement value can also be specified.

```python
cleaned_arr1 = np.nan_to_num(arr)
print(cleaned_arr1)

cleaned_arr2 = np.nan_to_num(arr, nan=100)
print(cleaned_arr2)
```

## Detecting Infinite Values with isinf()

- Use isinf() to identify positive and negative infinite values.
- The function returns True where an infinite value is present.

```python
arr = np.array([1,2,np.inf,3,4,-np.inf,5])

print(np.isinf(arr))
```

## Replacing Infinite Values with nan_to_num()

- Use nan_to_num() to replace positive and negative infinite values.
- Specify custom values using the posinf and neginf parameters.

```python
cleaned_arr = np.nan_to_num(arr, posinf=1000, neginf=-1000)

print(cleaned_arr)
```
