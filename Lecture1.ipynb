# PYTHON LISTS VS NUMPY ARRAYS

```python
import numpy as np
import time
import sys
```

```python
# EXECUTION TIME COMPARISON
size = 100_00_00

# PYTHON LISTS
python_list = range(size)
start = time.time()
square_list = [i*i for i in python_list]
end = time.time()
print(f"EXECUTION TIME OF PYTHON LISTS : {end-start} SECONDS")

# NUMPY ARRAYS
arr = np.array(python_list)
start = time.time()
square_array = arr ** 2
end = time.time()
print(f"EXECUTION TIME OF NUMPY ARRAYS : {end-start} SECONDS")
```

```python
# MEMORY USAGE COMPARISON

# PYTHON LISTS
print(f"SIZE OF PYTHON LISTS : {sys.getsizeof(python_list) * len(python_list)} BYTES")

# NUMPY ARRAYS
print(f"SIZE OF NUMPY ARRAYS : {arr.nbytes} BYTES")
```
