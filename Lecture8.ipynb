# NumPy Mathematics, Linear Algebra, and Statistics

- Learn how to perform mathematical operations on NumPy arrays.
- Apply trigonometric functions to array elements.
- Explore common linear algebra operations such as matrix multiplication, determinant, inverse, and trace.
- Use statistical and aggregate functions to analyze numerical data.

## Mathematical Operations on Arrays

- Perform arithmetic operations directly on NumPy arrays.
- NumPy applies operations element-wise to all values in the array.

```python
import numpy as np
```

```python
a = np.array([2,4,6,8,10])

print("a =", a)
print("a + 2 =", a + 2)
print("a - 2 =", a - 2)
print("a * 2 =", a * 2)
print("a / 2 =", a / 2)
print("a ** 2 =", a ** 2)
```

## Array-to-Array Arithmetic Operations

- Perform element-wise operations between arrays of the same shape.
- Corresponding elements from each array are combined.

```python
b = np.array([1,2,3,4,5])

print("b =", b)
print("a + b =", a + b)
```

## Trigonometric Functions

- Apply trigonometric functions to every element of an array.
- NumPy performs the calculations element-wise.

```python
print(np.sin(a))
print(np.cos(a))
```

## Matrix Multiplication

- Use matmul() to perform matrix multiplication.
- The number of columns in the first matrix must match the number of rows in the second matrix.

```python
a = np.ones((2,3))
print(a)

b = np.full((3,2), 2)
print(b)

print(np.matmul(a,b))
```

## Determinant of a Matrix

- The determinant is a scalar value associated with a square matrix.
- It is commonly used in linear algebra to determine matrix properties.

```python
c = np.identity(3)

d = np.array([
    [11,2,3],
    [4,15,16],
    [17,8,19]
])

print(np.linalg.det(c))
print(np.linalg.det(d))
```

## Inverse of a Matrix

- Use inv() to calculate the inverse of a square matrix.
- A matrix must be non-singular to have an inverse.

```python
f = np.array([
    [11,2,3],
    [4,15,16],
    [17,8,19]
])

print(np.linalg.inv(f))
```

## Trace of a Matrix

- The trace of a matrix is the sum of its main diagonal elements.
- Use trace() to calculate this value.

```python
t = np.array([
    [11,2,3],
    [4,15,16],
    [17,8,19]
])

print(np.trace(t))
```

## Statistical and Aggregate Functions

- Use aggregate functions to summarize numerical data.
- Common measures include minimum, maximum, mean, sum, standard deviation, and variance.

```python
stats = np.array([
    [1,2,3],
    [4,5,6]
])

print(stats)

print("MAX =", np.min(stats))
print("MIN =", np.max(stats))
print("MEAN/AVERAGE =", np.mean(stats))
print("SUM =", np.sum(stats))
print("STANDARD DEVIATION =", np.std(stats))
print("VARIANCE =", np.var(stats))
```

## Row-Wise and Column-Wise Operations

- Use the axis parameter to perform calculations across rows or columns.
- axis=0 performs column-wise operations.
- axis=1 performs row-wise operations.

```python
print("ROW-WISE MAX =", np.max(stats, axis=1))
print("COLUMN-WISE MAX =", np.max(stats, axis=0))

print("ROW-WISE MIN =", np.min(stats, axis=1))
print("COLUMN-WISE MIN =", np.min(stats, axis=0))

print("ROW-WISE SUM =", np.sum(stats, axis=1))
print("COLUMN-WISE SUM =", np.sum(stats, axis=0))
```
